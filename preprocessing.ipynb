```python
#importing libraries#i

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing dataset

ds=pd.read_csv('mental_health_data.csv')
ds
```

```python
ds.drop(columns=['User_ID'], inplace=True)
```

```python
#handling missing values
ds['Severity'].fillna(ds['Severity'].mode()[0], inplace=True)

#simple encoding 
ds['Severity'] = ds['Severity'].replace({
    'None': 0,
    'Low': 1,
    'Medium': 2,
    'High': 3
})
```

```python
#encoding on column stress level

ds['Stress_Level'] = ds['Stress_Level'].replace({
    'Low': 1,
    'Medium': 2,
    'High': 3
})
```

```python
ds['Stress_Level'].unique()
```

```python
#encoding diet quality column


ds['Diet_Quality'] = ds['Diet_Quality'].replace({
    'Unhealthy': 1,
    'Average': 2,
    'Healthy': 3
})
```

```python
ds['Consultation_History'] = ds['Consultation_History'].replace({
    'No': 0,
    'Yes': 1
})
```

```python
#encoding meditation usage column
ds['Medication_Usage'] = ds['Medication_Usage'].replace({'No': 0, 'Yes': 1})
```

```python
def remove_stop_words(text):
    #get the english stopword lists
    stop_words=set(stopwords.words('english'))
    #tokenize the sentence
    words=text.split(',')
    #remove stopwords
    filtered_words=[word for word in words if word.lower() not in stop_words]
    ##convert back to string 
    filtered_text =",".join(filtered_words)
    return filtered_text
```
